```python
import requests
from bs4 import BeautifulSoup

def scrape_news(category):
    base_url = 'https://www.viva.co.id/berita/'
    category_url = base_url+category
    # Lakukan request ke halaman web menggunakan requests
    response = requests.get(category_url)

    # Periksa apakah request berhasil
    if response.status_code == 200:
        # Gunakan BeautifulSoup untuk parsing halaman web
        soup = BeautifulSoup(response.content, 'html.parser')

        # Temukan elemen-elemen yang berisi berita pada halaman web

        news_articles = soup.find_all('div', class_='article-list-row')  # Sesuaikan dengan struktur HTML dari situs yang ingin di-scrape
        data_berita = []
        data_link = []
        # Loop melalui artikel berita dan ambil informasi yang dibutuhkan
        for article in news_articles:
            # Misalnya, ambil judul dan link artikel
            headline = article.find('a', class_='article-list-title').text.strip()
            data_berita.append(headline)
            article_link = article.find('a')['href']
            data_link.append(article_link)
        return data_berita
    else:
        print("Gagal melakukan request")

# Contoh penggunaan:
# Ganti 'https://example.com/category' dengan URL kategori berita yang ingin Anda scrape
title_article = []
kategori = ['nasional', 'politik', 'kriminal']
for cat in kategori:
  result_judul = scrape_news(cat)
  title_article.extend(result_judul)
```
