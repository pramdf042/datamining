```python
#mengambil data
dataset_url= "https://raw.githubusercontent.com/pramdf042/datamining/main/data-penindakan-pelanggaran-lalu-lintas-dan-angkutan-jalan-tahun-2021-bulan-januari.csv"
p=pd.read_csv(dataset_url,index_col="wilayah")
```

```python
p
```

```python
def diss(x,y):
  a = 8
  b = 0
  for col in ["bap_tilang","stop_operasi","bap_polisi","stop_operasi_polisi","penderekan","ocp_roda_dua","ocp_roda_empat","angkut_motor"]:
    if p[col][x] == p[col][y]:
      b+=1
    return (a-b)/a
```

```python
print(diss(1,2))
print(diss(3,3))
print(diss(4,5))
```

```python
p.describe()
```

```python
#normalisasi rating
std_bapt = 74.580605
mean_bapt = 112.666667
for col1 in ["bap_tilang"]:
    p1 = p[col1]
    hasil1 = (p1-mean_bapt)/std_bapt
    print(hasil1)

#normalisasi metascore
std_stopo = 64.450498	
mean_stopo = 63.333333
for col2 in ["stop_operasi"]:
    p2 = p[col2]
    hasil2 = (p2-mean_stopo)/std_stopo
    print(hasil2)
```

```python
# Menghitung jarak data Numeric dengan cara Euclidean
def num_p(a,b,c):
    result = 0
    for col in [hasil1, hasil2]:
        result = math.pow(abs(col[a] - col[b]), c)
    return result**(1/c)
```

```python
print(num_p(1,2,2))
print(num_p(1,3,2))
print(num_p(1,4,2))
```

```python
print("Jarak d(1,2) =", diss(1,2)+num_p(1,2,2))
```

```python
print("Jarak d(3,3) =", diss(3,3)+num_p(1,3,2))
```

```python
print("Jarak d(2,5) =", diss(2,5)+num_p(1,4,2))
```
